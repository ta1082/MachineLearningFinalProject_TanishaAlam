The goal of this project is to classify 6 different breast tissues that are related to cancer.
The six different breast cancer tissue are:
Caricinoma FibroAdenoma Masopathy Gladular Connective Adipose.

The dataset and the paper by Silva & Marques can be downloaded by clicking on the link below:
https://archive.ics.uci.edu/ml/datasets/Breast+Tissue

There are 6 different classes and 9 features.

Next I have chosen two features from the data: I0 and PA500 which is the Impedivity at Zero Frequency and Phase Angle at 500 Khz to see a correlation of the data.

I plotted each tissue corresponding to the two features and the results are produced below.

```python
yraw=np.array(df['Class'])
xraw= ['I0', 'PA500']
twoValues=np.array(df[xraw])
import pylab as pl
Caricinoma= (yraw == 'car')
FibroAdenoma= (yraw == 'fad')
Masopathy= (yraw == 'mas')
Gladular=(yraw == 'gla')
Connective= (yraw == 'con')
Adipose= (yraw == 'adi')


plt.plot
plt.plot(twoValues[Caricinoma,0], twoValues[Caricinoma,1], 'r.')
plt.plot(twoValues[FibroAdenoma,0], twoValues[FibroAdenoma,1], 'b.')
plt.plot(twoValues[Masopathy,0], twoValues[Masopathy,1], 'g.')
plt.plot(twoValues[Gladular,0], twoValues[Gladular,1], 'c.')
plt.plot(twoValues[Connective,0], twoValues[Connective,1], 'y.')
plt.plot(twoValues[Adipose,0], twoValues[Adipose,1], 'k.')
plt.xlabel('Impedivity (Ohm) At Zero Frequency')
plt.ylabel('Phase Angle at 500 Khz')
plt.legend(['Caricinoma', 'Fibro-Adenoma', 'Masopathy', 'Gladular', 'Connective', 'Adipose'], loc= 'upper right')

```

Next I used sklearn's SVC svm classifier to classify the type of breast tissue for all 6 classes.
I had to scale the features because some values would be as small as 0.2 and large as 18000.

I had to randomly permutate the data because the dataset had all the breast tissues grouped together one after another.

```python
from sklearn import svm
xraw2= ['I0', 'PA500', 'HFS', 'DA', 'Area', 'A/DA', 'Max IP', 'DR', 'P' ]
allFeatures= np.array(df[xraw2])
```

```python
from sklearn.preprocessing import scale
X = scale(allFeatures)
```

```python
numDataPoints=yraw.shape[0]
ntr=80;
nts=numDataPoints-26;

Iperm=np.random.permutation(106)
Xtr= X[Iperm[:ntr],:]
ytr=yraw[Iperm[:ntr]]
Xts= X[Iperm[ntr:ntr+nts], :]
yts= yraw[Iperm[ntr:ntr+nts]]
```

Using the kernel rbf, C=200 and gamma=0.02 I predicted the test data on the svm classifier.

```python
from sklearn import svm
svc= svm.SVC(probability=False, kernel='rbf', C=200, gamma=0.02, verbose=10)
svc.fit(Xtr,ytr)
```

```python
yhat= svc.predict(Xts)
acc= np.mean(yhat == yts)
print('The accuracy for the SVM classifier is = {0:f}'.format(acc))
```

Next realizing that the accuracy isn't as high as I'd like it to be I decided to perform a KFold of 5 and using cross validation in order to optimize for the highest accuracy.
I looped over different C and gamma values in the logarithimic scale.

```python
from sklearn.model_selection import KFold
nfold=5
kf=KFold(n_splits=nfold,shuffle=True)
CTest = np.logspace(0,3,10)
gammaTest = np.logspace(-3,0,10)
ngam = len(gammaTest)
nC = len(CTest)
CData= np.zeros((nfold,ngam,nC))
gammaData=np.zeros((nfold,ngam,nC))
yhatData=np.zeros((nfold,ngam,nC))
accData=np.zeros((nfold,ngam,nC))

for ifold, I in enumerate(kf.split(X)):
    train, test = I
    Xtr= X[train,:]
    ytr= yraw[train]
    Xts= X[test, :]
    yts= yraw[test]
    
    
    from sklearn import svm
    for i, C in enumerate(CTest):
        for j, gamma in enumerate(gammaTest):
            svc= svm.SVC(probability=False, kernel='rbf', C=C, gamma=gamma, verbose=0)
            svc.fit(Xtr,ytr)
            yhat=svc.predict(Xts)
            acc= np.mean(yhat == yts)
            
    
            CData[ifold,j,i] = C
            gammaData[ifold,j,i] = gamma
            accData[ifold,j,i] = acc
            print('ifold=%d C=%12.4e gam=%12.4e acc=%12.4e' % (ifold,C,gamma,acc))



    
    

    
```

Below I have calculated the average accuracy throughout all the folds.
They are in the mid 70s range, which isn't so much different than using C=200 and gamma=0.02.
However when I got the maximum accuracy generated over the folds it was in the 80-90% range.

```python
acc_avg = np.mean(accData, axis=0)
plt.imshow(acc_avg)
plt.colorbar()
maxAccuracy=np.max(accData)
print("The max accuracy over all the folds was {0:.4f}".format(maxAccuracy))
```
